# NMSI Two-Flux 1D Demo (Notebook)

This notebook visualizes the 1D two-flux demo (active A, passive P) with NMSI operators.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('twoflux_nmsi_1d_timeseries.csv')
df.head()
```

```python
plt.figure();
plt.plot(df['x'], df['u_A'], label='u_A (active)');
plt.plot(df['x'], df['u_P'], label='u_P (passive)');
plt.plot(df['x'], df['w_rel'], label='w');
plt.legend(); plt.grid(True); plt.xlabel('x'); plt.ylabel('velocity');
plt.title('Velocities and Relative Mismatch');
```

```python
plt.figure();
plt.plot(df['x'], df['E'], label='Energy proxy');
plt.plot(df['x'], df['stress'], label='Stress (CME proxy)');
plt.plot(df['x'], df['gamma_diss'], label='gamma_diss');
plt.legend(); plt.grid(True); plt.xlabel('x'); plt.ylabel('arb.');
plt.title('Energy vs Stress & Selective Dissipation');
```

```python
plt.figure();
plt.plot(df['x'], df['theta_deg'], label='theta (deg)');
plt.plot(df['x'], df['brightness'], label='brightness');
plt.plot(df['x'], df['bifurcation_idx'], label='bifurcation');
plt.legend(); plt.grid(True); plt.xlabel('x'); plt.ylabel('observables');
plt.title('Observables: Deflection, Brightness, Bifurcation');
```
