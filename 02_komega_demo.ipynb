# k–ω (toy) driven by NMSI operators


```python
import numpy as np
import matplotlib.pyplot as plt
from nmsi_glider_ext.core.nmsi_turbulence import komega_nmsi
ts = np.linspace(0,20,120)
res = [komega_nmsi(7.0, t, 5.0) for t in ts]
q0 = [r['q_base'] for r in res]; qN = [r['q_nmsi'] for r in res]; qT = [r['q_turb'] for r in res]
plt.plot(ts,q0,label='Base'); plt.plot(ts,qN,label='NMSI'); plt.plot(ts,qT,label='k-ω (toy)');
plt.legend(); plt.grid(True); plt.xlabel('t [s]'); plt.ylabel('qdot (arb)');
```
