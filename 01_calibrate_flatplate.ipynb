# Calibration — Flat Plate, Mach 7
Compare baseline vs NMSI-modulated heat flux.


```python
import yaml
from core.operators import nmsi_glider_step
with open('../cases/flat_plate_M7.yaml') as f:
    cfg = yaml.safe_load(f)
M, aoa = cfg['mach'], cfg['aoa_deg']
rho, a, V = cfg['rho_inf'], cfg['a_sound'], cfg['a_sound']*cfg['mach']
Rn, dcp, divu = cfg['Rn'], cfg['dCp_dx'], cfg['div_u']
params, Cf_mod = cfg['params'], cfg['Cf_mod']
ops, q0, qN, gain = nmsi_glider_step(M, 10.0, aoa, rho, V, Rn, dcp, divu, params, Cf_mod)
print('qdot_base =', q0)
print('qdot_nmsi =', qN)
print('gain =', gain)
```
