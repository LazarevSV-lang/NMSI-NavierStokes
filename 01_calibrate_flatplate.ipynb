# Calibration — Flat Plate, Mach 7
Load case, run NMSI operators, compare qdot_base vs qdot_nmsi.


```python
import yaml, numpy as np
from core.operators import nmsi_glider_step

with open('../cases/flat_plate_M7.yaml') as f:
    cfg = yaml.safe_load(f)

M = cfg['mach']; aoa = cfg['aoa_deg'];
rho = cfg['rho_inf']; a = cfg['a_sound']; V = M*a
Rn = cfg['Rn']; dcp = cfg['dCp_dx']; divu = cfg['div_u']
params = cfg['params']; Cf_mod = cfg['Cf_mod']

ops, q0, qN, gain = nmsi_glider_step(M, 10.0, aoa, rho, V, Rn, dcp, divu, params, Cf_mod)
print('qdot_base =', q0, 'W/m^2')
print('qdot_nmsi =', qN, 'W/m^2')
print('gain      =', gain)
```
