# Quick Visualization — Energy / Enstrophy / Max Vorticity
Run the 2D demo first:
```bash
python nmsi_nse2d_demo.py --config config.yaml
```
It creates `out_2d_diag.csv`. Then run the cells below.

````python
import csv
import matplotlib.pyplot as plt
t,E,Om,Wmax=[],[],[],[]
with open('out_2d_diag.csv','r') as f:
    r=csv.DictReader(f)
    for row in r:
        t.append(float(row['t'])); E.append(float(row['E']))
        Om.append(float(row['Omega'])); Wmax.append(float(row['Wmax']))
# Energy
plt.figure(); plt.plot(t,E); plt.xlabel('t'); plt.ylabel('Energy E(t)'); plt.title('Energy vs Time'); plt.show()
# Enstrophy
plt.figure(); plt.plot(t,Om); plt.xlabel('t'); plt.ylabel('Enstrophy Ω(t)'); plt.title('Enstrophy vs Time'); plt.show()
# Max vorticity
plt.figure(); plt.plot(t,Wmax); plt.xlabel('t'); plt.ylabel('||ω||_∞'); plt.title('Max Vorticity vs Time'); plt.show()
````
